```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('../data/municipios.xlsx')
```

```python
df.head(50)
```

```python
# Create new columns for 'ESTADO' and 'MUNICIPIO'
df['ESTADO'] = np.nan
df['MUNICIPIO'] = np.nan

# Create a mask where 'Unnamed: 0' is NaN
mask = df['Unnamed: 0'].isna()

# Set 'ESTADO' to 'AGUASCALIENTES' where 'mask' is True
df.loc[mask, 'ESTADO'] = df.loc[mask, 'AGUASCALIENTES']

mask = df['Unnamed: 0'].notna()

# Set 'MUNICIPIO' to 'AGUASCALIENTES' where 'mask' is True
df.loc[mask, 'MUNICIPIO'] = df.loc[mask, 'AGUASCALIENTES']
```

```python
df = df.drop(columns='Unnamed: 0')
```
